```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
class stoch_dyn_bubble:
    def __init__(self, _m, _a, _b, _mu, _sigma):
        self.m = _m
        self.a = _a
        self.b = _b
        self.mu = _mu
        self.sigma = _sigma
        self.bval = 0.101
        self.bval_last = 0.1
        
        self.mu0_tot = 0
        self.time_steps = 0
    
    def get_value(self):
        
        nu = random.gauss(self.mu,self.sigma)
        mu0 = (self.bval-self.bval_last)/self.bval_last
        
        self.bval_last = self.bval
        alph = self.a*mu0
        beta = self.b*nu
        self.bval += (alph+beta)*self.bval**(self.m)
        
        self.mu0_tot += mu0
        self.time_steps += 1
        
        return [self.bval,alph,beta]
    
    def get_consts(self):
        return [self.m,self.a,mu0,self.b]
    
    def get_bubble_info(self):
        return [self.time_steps, self.mu0_tot]
```

```python
class ParamError(OverflowError):
    pass

class stock_history:
    
    def __init__(self, months, *params):
        self.months = months
        params = params[0]
        self.B = stoch_dyn_bubble(*params)
        try:
            self.minutes = range(60*24*30*months)
            self.stock_dat = [self.B.get_value() for minute in self.minutes]
            self.stock_dat = zip(*self.stock_dat)
            print 'max/min return rates (rate/min):', round(max(self.stock_dat[1]),6),'/',round(min(self.stock_dat[1]),6)
        except:
            raise ParamError('environment parameters may be unstable. reset them or try again.')
    
    def plot_history(self, limits=None, savefile = None):
        
        if limits!=None:
            tmin = limits[0]
            tmax = limits[1]
            thickness = 4/(tmax-tmin)
            x= self.minutes[int(tmin*43200):int(tmax*43200)]
            y0 = self.stock_dat[0][int(tmin*43200):int(tmax*43200)]
            y1 = self.stock_dat[1][int(tmin*43200):int(tmax*43200)]
            y2 = self.stock_dat[2][int(tmin*43200):int(tmax*43200)]
        else:
            thickness = 1
            x= self.minutes
            y0 = self.stock_dat[0]
            y1 = self.stock_dat[1]
            y2 = self.stock_dat[2]


        my0 = max(abs(np.array(y0)))
        my1 = max(abs(np.array(y1)))
        my2 = max(abs(np.array(y2)))

        fig,ax = plt.subplots(figsize = (30,15))
        params = {'legend.fontsize': 20,
                  'legend.linewidth': 1.5,
                  'legend.markerscale': 1}
        plt.rcParams.update(params)

        plt.plot(x,y0,linewidth = 0.8, color = 'k')
        plt.plot(x,np.array(y2)*my0/my2/2, 'r-', linewidth = 0.015*thickness)
        plt.plot(x,np.array(y1)*my0/my1/2, 'b-', linewidth = 0.02*thickness)
        plt.ylim(1.05*min(y1)*my0/my1/1.5, 1.05*max(y0))
        plt.plot(0,0,'k-', label = 'Stock Price (B)')
        plt.plot(0,0,'b-', label = 'dB/dt (scaled)')
        plt.plot(0,0,'r-', label = 'Gaussian White Noise (scaled)')
        plt.legend(loc = 'best')
        
        plt.xlim(xmax = x[-1])
        
        plt.yticks(fontsize = 17)
        plt.xticks(x[::43200],fontsize = 17)
        if limits!=None:
            ax.set_xticklabels(range(tmin,tmax))
            plt.xlim(tmin*43200,tmax*43200)
        else:
            ax.set_xticklabels(range(0,self.months))
        plt.xlabel('Months', fontsize = 25)
        
        if savefile!=None:
            plt.savefig(savefile, bbox_inches = 'tight')
        
        plt.show()
```

```python
#Suggested parameters: [2,0.045,0.01,0.004,0.7]
#The above parameters are not stable past 6 months
#See the saved_output directory for example output
Data = stock_history(6,[2,0.045,0.01,0.004,0.7])
```
